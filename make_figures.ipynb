```python
import contextily as cx
import geopandas as gpd
import re
import os
```

```python
def get_files(ext, direc):
    return [i for i in os.listdir(direc) if i.endswith(ext)]
```

```python
shps = []
base_dir = "shps"

for shp in os.listdir(base_dir):
    if ".ipynb" not in shp:
        fname = os.path.join(base_dir, shp, get_files(".shp", os.path.join(base_dir, shp))[0]) 
        print(fname)
        shps.append(gpd.read_file(fname))
```

```python
adm0 = gpd.pd.concat(shps)
adm0 = adm0.to_crs(epsg=3857)
```

```python
ax = adm0.plot(figsize=(10, 10), alpha=0.5, edgecolor='k')
cx.add_basemap(ax, source = cx.providers.OpenStreetMap.Mapnik)
```
